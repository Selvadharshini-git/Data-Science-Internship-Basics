```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Display the first 5 rows
print("First 5 rows of the dataset:")
print(df.head())

# Check for missing values
print("\nMissing values:")
print(df.isnull().sum())

# Generate summary statistics
print("\nSummary statistics:")
print(df.describe())

# Plot histograms
df.hist(figsize=(10, 8))
plt.suptitle('Histograms of Iris Dataset Features')
plt.show()

# Create a pairplot
sns.pairplot(df, hue='species')
plt.suptitle('Pairplot of Iris Dataset by Species', y=1.02)
plt.show()
```

Step 1: The dataset is loaded into a pandas DataFrame. The species column is added to the DataFrame to represent the target variable

Step 2: missing values in each column. If there are any, they need to be handled

Step 3:The describe() function provides key statistics such as mean, standard deviation, minimum, maximum, and quartiles for each numerical column.

Step 4:
 1.creates histograms for each feature (sepal length, sepal width, petal length, petal width)

2. pairplot visualizes pairwise relationships between features, with points colored by species. This helps identify patterns and correlations.

Step 5:
 1.Sepal Length and Width: Setosa has smaller sepal lengths but larger sepal widths compared to the other species.

2.Petal Length and Width: Virginica has the largest petal measurements, while setosa has the smallest.

3.Correlations: Petal length and petal width are highly correlated, as seen in the pairplot.

4.Distributions: The distributions of petal measurements are more distinct across species compared to sepal measurements.
