```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Display the first few rows of the dataset
print(iris.head())

# Create histograms for all numerical columns
iris.hist(figsize=(10, 8))
plt.suptitle('Histograms of Numerical Columns in Iris Dataset')
plt.show()

# Create box plots for all numerical columns
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris)
plt.title('Box Plots of Numerical Columns in Iris Dataset')
plt.show()

# Calculate the correlation matrix (excluding the 'species' column)
corr = iris.drop(columns=['species']).corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Iris Dataset')
plt.show()
```
